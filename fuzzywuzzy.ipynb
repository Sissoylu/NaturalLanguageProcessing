```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
fuzz.ratio("doğal dil işleme", "doğal dil işleme???")
```

```python
fuzz.ratio("bilet almak istiyorum", "yarın için bilet alacağım")
```

```python
#noktalama işaretlerinden etkilenmez
fuzz.partial_ratio("Natural Language Processing","Natural Language Processing???")
```

```python
fuzz.partial_ratio("bilet alacağım","bilet almaya gidiyorum")
```

```python
fuzz.token_set_ratio("doğal dil işleme", "doğal dil işleme işleme") 
```

```python
fuzz.token_set_ratio("bilet alacağım", "bilet almak istiyorum") 
```

```python
cumle= 'doğal dil'
liste = ['dil işleme doğal', 'doğal dil işlemek', 'dili işlemek']  
process.extract(cumle, liste) 
```

```python
cumle1 = "bilet almak"
liste1 = ['yarın bilet almaya gideceğim', 'biletçiye gideceğim', 'ankara gitmek için bilet alacağım']
process.extract(cumle1,liste1)
```

```python
#büyük küçük harflere ve noktalama işaretlerine takılmaz
fuzz.WRatio('doğal dil işlemek', 'Doğal Dil İşlemek') 
```
